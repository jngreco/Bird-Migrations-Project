```python
bird_df = pd.read_csv("birddata.csv")
```

```python
bird_df = bird_df.reset_index()
```

```python
bird_df.columns
```

```python
clean_df = bird_df[["COMMON NAME","SCIENTIFIC NAME","OBSERVATION COUNT","COUNTRY","STATE","COUNTY","LOCALITY","LATITUDE","LONGITUDE","OBSERVATION DATE","TIME OBSERVATIONS STARTED","PROTOCOL TYPE","PROTOCOL CODE","REVIEWED"]]
```

```python
bird_df["OBSERVATION COUNT"].value_counts()
```

```python
clean_df
```

```python
clean_df["OBSERVATION COUNT"] = clean_df["OBSERVATION COUNT"].replace("X",1)
```

```python
clean_df["OBSERVATION COUNT"] = clean_df["OBSERVATION COUNT"].astype(int)
```

```python
year = []
month = []
for date in clean_df["OBSERVATION DATE"]:
    year.append(int(date.split("-")[0]))
    month.append(int(date.split("-")[1]))
clean_df["OBSERVATION YEAR"] = year
clean_df["OBSERVATION MONTH"] = month
clean_df
```

```python
def generate_year_plot(bird):
    filter_df = clean_df.loc[clean_df["COMMON NAME"] == bird]
    gf_df = filter_df.groupby(["OBSERVATION YEAR"]).sum()
    plt.figure(figsize=(16,8))
    gf_df["OBSERVATION COUNT"].plot(kind="bar")
    plt.title(f"{bird} Sightings by Year")
    plt.ylabel("Recorded Sightings")
    plt.xlabel("Observation Year")
    plt.show()
    return
```

```python
def generate_month_plot(bird):
    filter_df = clean_df.loc[clean_df["COMMON NAME"] == bird]
    gf_df = filter_df.groupby(["OBSERVATION MONTH"]).sum()
    plt.figure(figsize=(8,6))
    gf_df["OBSERVATION COUNT"].plot(kind="bar")
    plt.title(f"{bird} Sightings by Month")
    plt.ylabel("Recorded Sightings")
    plt.xlabel("Observation Month")
    plt.show()
    return
```

```python
clean_df["COMMON NAME"].unique()
```

```python
generate_month_plot("Double-crested Cormorant")
```
